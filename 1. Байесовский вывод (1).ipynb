# Задача 1

Пусть $X_1, X_2, \ldots, X_n$ — выборка из экспоненциального распределения с параметром $\lambda$. Найти оценку максимального правдоподобия параметра $\lambda$, сравнить ее с байесовской оценкой (MAP и математическое ожидание апостреорного распределения), подобрав сопряженное распределение. Сравнить полученные байесовские оценки с оценкой MLE. Найти предсказательное распределение

<p>Решение: </p>
<p>$X_1,X_2,..., X_n ≃ X$ из $Exp(λ)$</p>
<p>1. Правдоподобие и MLE оценка </p>
<p>$P(X|λ) = P(X_1 < x) * ... * P(X_n < x) = λ^ne^{-λ(X_1 + ... + X_n)}$</p>
<p>$(P(X|λ))$ достигает максимума при $ λ_{mle} <=> (ln P(X|λ_{mle}))' = 0$</p>
<p>$(ln P(X|λ_{mle}))' = \frac{n}{λ_{mle}} - (X_1 + ... + X_n) = 0 => λ_{mle} = \frac1{mean(X)}$</p>
<p>2. Априорное распределение</p>
<p>Сопряженное к экспоненциальному это гамма с параметрами $\alpha, \beta$ :</p>
<p>$P(λ) = \frac{{(\beta + (X_1 + ... + X_n))}^{\alpha + n}}{Г(\alpha + n)}*λ^{α-1}*e^{-λβ}$</p>
<p>3. Апостериорное распределение</p>
<p>$P(λ|X) ∝ P(X|λ)*P(λ) ∝ \frac{{(\beta + (X_1 + ... + X_n))}^{\alpha + n}}{Г(\alpha + n)}λ^ne^{-λ(X_1 + ... + X_n)}*λ^{α-1}*e^{-λβ}$</p>
<p>$P(λ|X) ≃ Gamma(α + n, β + (X_1 + ... + X_n))$</p>
<p>4. Среднее апостериорного распределения</p>
<p>$λ_{bayes} = E[P(λ|X)] = \frac{α + n}{β + (X_1 + ... + X_n)}$</p>
<p>Получаем $λ_{mle}$, когда α и β стремятся к 0.</p>
<p>5. MAP (мода апостериорного распределения)</p>
<p>$(ln(P(λ|X))' = \frac{\alpha + n - 1}{λ_{map}} - \beta - (X_1 + ... + X_n) = 0 => λ_{map} = \frac{α + n - 1}{β + (X_1 + ... + X_n)}$</p>
<p>Получили то же самое, но смещенное на 1.</p>
<p>6. Предсказательное распределение</p>
<p>$P(X_{n+1}|X) = \int_0^{+∞} P(X_{n+1}|λ)P(λ|X)\, dλ = \int_0^{+∞} λ*e^{-λX_{n+1}}*\frac{{(\beta + (X_1 + ... + X_n))}^{\alpha + n}}{Г(\alpha + n)}λ^ne^{-λ(X_1 + ... + X_n)}*λ^{α-1}*e^{-λβ}\, dλ = \frac{{(\beta + (X_1 + ... + X_n))}^{\alpha + n}}{Г(\alpha + n)} * \frac{Г(\alpha + n + 1)}{(\beta + (X_1 + ... + X_n + X_{n+1}))^{\alpha + n + 1}} = \frac{(\alpha + n){(\beta + (X_1 + ... + X_n))}^{\alpha + n}}{(\beta + (X_1 + ... + X_n + X_{n+1}))^{\alpha + n + 1}}$</p>




# Задача 2

**Мультиномиальное распределение**

Пусть проводится серия из $n$ испытаний и в результате каждого испытания происходит ровно одно событие из набора $A_1, A_2, \dots, A_m$, причем вероятности этих событий равны соответственно $\mathsf{p}_1, \mathsf{p}_2, \dots, \mathsf{p}_m$, причем
$$\sum_{i=1}^{m}\mathsf{p}_i = 1.$$

Тогда совместное распределение величин $X_1, X_2, \dots, X_m$, где $X_k$ — число наступлений события $A_k$ в серии из $n$ испытаний, задается вероятностями

$$
\mathsf{P}\left(X_1 = n_1, \dots, X_m = n_m, \right) = \frac{n!}{n_1!\dots n_m!}\mathsf{p}_1^{n_1}\dots \mathsf{p}_m^{n_m},
$$

где $n_1, n_2, \dots, n_m$ — произвольный набор целых неотрицательных чисел, таких что

$$\sum_{i=1}^m n_i = n.$$

Произведите байесовский вывод для мультиномиального распределения: найдите апостериорное распределение, используя в качестве сопоряженного распределения к правдоподобию [распределение Дирихле](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%94%D0%B8%D1%80%D0%B8%D1%85%D0%BB%D0%B5), найдите предсказательное распределение. Объясните результат.

<p>Решение:</p>
<p>1. Правдоподобие</p>
<p>$ \theta = p_1, ..., p_m $</p>
<p>$ X = X_1, ..., X_m $</p>
<p>$P(X|\theta) = \frac{n!*p_1^{n_1}*...*p_m^{n_m}}{n_1!*...n_m!}$</p>
<p>2. Априорное распределение</p>
<p>$P(\theta) = \frac{p_1^{\alpha_1 - 1}*...*p_m^{\alpha_m - 1}}{Β(α)} ≃ Dir(\alpha_1,...,\alpha_m)$</p>
<p>3. Апостериорное распределение</p>
<p>$P(\theta|X) ∝  P(X|\theta)*P(\theta) ∝ p_1^{\alpha_1 + n_1 - 1}*...*p_m^{\alpha_m + n_m - 1}$</p>
<p>$P(\theta|X) ≃ Dir(\alpha_1 + n_1,...,\alpha_m + n_m)$</p>
<p>4. Байесовская оценка</p>
<p>$E[P(p_i|X)] = \frac{\alpha_i + n_i}{\alpha_1 + n_1 + ... + \alpha_m + n_m}$</p>
<p>5. Предсказательное распределение</p>
<p>$P(X_{m+1} = k|X) = ∫P(X_{m+1} = k|\theta)P(\theta|X)d\theta ∝ ∫p_k*p_1^{\alpha_1 + n_1 - 1}*...*p_m^{\alpha_m + n_m - 1}d\theta = E[P(p_k|X)] = \frac{\alpha_k + n_k}{\alpha_1 + n_1 + ... + \alpha_m + n_m}$</p>
<p>6. Выводы</p>
<p>В качестве байесовской оценки и предсказания, как и в примере с обычной монеткой, рассмотренном на лекциях, получили сглаженную относительную частоту события $A_i$ с учетом смещения нашей "многомерной монетки" (для моделирования которой мы используем распределение Дирихле, то есть многомерное бэта распределение).</p>
